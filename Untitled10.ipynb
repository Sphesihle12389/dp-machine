```python
# 3) Heuristics to map columns to expected fields
# Expected fields: member_id, obs_date/start_date, age, salary, tenure, balance_access, balance_retire, withdraw_event, withdraw_date/time_days
col_lower = {c: c.lower().strip() for c in df.columns}

def find_col_by_keywords(keywords):
    # returns first column name where any keyword is in lowercased column name
    for orig, low in col_lower.items():
        for kw in keywords:
            if kw in low:
                return orig
    return None

mapping = {}
mapping['member_id'] = find_col_by_keywords(['member', 'id', 'member_id', 'emp', 'employee'])
mapping['obs_date']   = find_col_by_keywords(['date', 'start', 'joined', 'join', 'obs_date', 'observation'])
mapping['age']        = find_col_by_keywords(['age'])
mapping['salary']     = find_col_by_keywords(['salary', 'pay', 'wage', 'income'])
mapping['tenure']     = find_col_by_keywords(['tenure', 'service', 'years', 'yrs'])
mapping['balance_access'] = find_col_by_keywords(['access', 'accessible', 'withdrawable', 'pot1', 'pot_1', 'pota', 'liquid'])
mapping['balance_retire'] = find_col_by_keywords(['retire', 'retirement', 'preserve', 'pot2', 'pot_2', 'potb', 'locked'])
mapping['withdraw_event']  = find_col_by_keywords(['withdraw', 'withdrawal', 'withdrew', 'attrition', 'left', 'exit', 'event'])
mapping['withdraw_date']   = find_col_by_keywords(['withdraw_date', 'withdrawal_date', 'exit_date', 'left_date', 'resign_date'])

print("\nAutomatic column mapping (best-effort):")
for k,v in mapping.items():
    print(f"  {k:16s} -> {v}")
```

```python

# 4) Preprocess / create expected columns if missing
df2 = df.copy()

# Member ID
if mapping['member_id'] is None:
    print("member_id not found; creating synthetic member_id from index.")
    df2['member_id'] = np.arange(1, len(df2)+1)
    mapping['member_id'] = 'member_id'
else:
    # rename for convenience to standard names
    df2.rename(columns={mapping['member_id']:'member_id'}, inplace=True)
    mapping['member_id'] = 'member_id'

# obs_date
if mapping['obs_date'] is None:
    print("obs_date not found; creating a synthetic 'start_date' using a uniform distribution from 2018-01-01 to 2021-12-31.")
    df2['start_date'] = pd.to_datetime('2018-01-01') + pd.to_timedelta(np.random.randint(0, 365*4, size=len(df2)), unit='d')
    mapping['obs_date'] = 'start_date'
else:
    df2.rename(columns={mapping['obs_date']:'start_date'}, inplace=True)
    mapping['obs_date'] = 'start_date'
    # try parse
    try:
        df2['start_date'] = pd.to_datetime(df2['start_date'])
    except Exception:
        print("Warning: could not parse start_date column to datetime; leaving as-is.")

# Age
if mapping['age'] is None:
    print("age not found; creating synthetic 'age' (25-55).")
    df2['age'] = np.random.randint(25, 55, size=len(df2))
    mapping['age'] = 'age'
else:
    df2.rename(columns={mapping['age']:'age'}, inplace=True)
    mapping['age'] = 'age'

# Salary
if mapping['salary'] is None:
    print("salary not found; creating synthetic 'salary' (3000-40000).")
    df2['salary'] = np.random.randint(3000, 40000, size=len(df2))
    mapping['salary'] = 'salary'
else:
    df2.rename(columns={mapping['salary']:'salary'}, inplace=True)
    mapping['salary'] = 'salary'

# Tenure
if mapping['tenure'] is None:
    print("tenure not found; creating synthetic 'tenure' (0-30).")
    df2['tenure'] = np.random.randint(0, 30, size=len(df2))
    mapping['tenure'] = 'tenure'
else:
    df2.rename(columns={mapping['tenure']:'tenure'}, inplace=True)
    mapping['tenure'] = 'tenure'

# balance_access
if mapping['balance_access'] is None:
    print("balance_access not found; creating synthetic 'balance_access' (0-40000).")
    df2['balance_access'] = np.random.randint(0, 40000, size=len(df2))
    mapping['balance_access'] = 'balance_access'
else:
    df2.rename(columns={mapping['balance_access']:'balance_access'}, inplace=True)
    mapping['balance_access'] = 'balance_access'

# balance_retire
if mapping['balance_retire'] is None:
    print("balance_retire not found; creating synthetic 'balance_retire' (10000-200000).")
    df2['balance_retire'] = np.random.randint(10000, 200000, size=len(df2))
    mapping['balance_retire'] = 'balance_retire'
else:
    df2.rename(columns={mapping['balance_retire']:'balance_retire'}, inplace=True)
    mapping['balance_retire'] = 'balance_retire'

# withdraw_event
if mapping['withdraw_event'] is None:
    # try to find columns indicating attrition-like signals
    print("withdraw_event not found. Creating synthetic 'withdraw_event' using simple risk model based on balances/tenure.")
    # synthetic probability: higher prob when balance_access small and tenure low
    bal = df2['balance_access'].fillna(0) + df2['balance_retire'].fillna(0)
    risk = 1 / (1 + np.exp((bal)/1e5 - 1)) + (10 - df2['tenure'].fillna(5)) / 20
    prob = 1/(1+np.exp(-(-3.0 + 0.00005*df2['balance_access'] + 0.05*(10 - df2['tenure']))))
    df2['withdraw_event'] = np.random.binomial(1, prob.clip(0,0.9))
    mapping['withdraw_event'] = 'withdraw_event'
else:
    df2.rename(columns={mapping['withdraw_event']:'withdraw_event'}, inplace=True)
    mapping['withdraw_event'] = 'withdraw_event'
    # normalize to 0/1 if not numeric
    if df2['withdraw_event'].dtype == object:
        df2['withdraw_event'] = df2['withdraw_event'].map(lambda x: 1 if str(x).strip().lower() in ['1','yes','y','true','t','withdraw','left','attrited'] else 0)

# withdraw_date / time -> time_days
if mapping['withdraw_date'] is None:
    print("withdraw_date not found; building 'time_days' either from an existing numeric time column or synthesizing with cap at 5 years.")
    # attempt to find any numeric column that could be time
    possible_time = None
    for col in df2.columns:
        if df2[col].dtype.kind in 'iuf' and 'time' in col.lower():
            possible_time = col
            break
    if possible_time:
        df2['time_days'] = df2[possible_time].astype(int)
        print("Using numeric column for time_days:", possible_time)
    else:
        # generate synthetic times: event -> exponential, else cap
        max_followup_days = 365*5
        # for those with withdraw_event==1, draw exponential; others set to cap
        idx_event = df2['withdraw_event'] == 1
        rand_time = np.random.exponential(scale=600, size=idx_event.sum()).astype(int)
        df2.loc[idx_event, 'time_days'] = rand_time
        df2.loc[~idx_event, 'time_days'] = max_followup_days
    mapping['withdraw_date'] = 'time_days'
else:
    df2.rename(columns={mapping['withdraw_date']:'withdraw_date'}, inplace=True)
    mapping['withdraw_date'] = 'withdraw_date'
    # try to coerce into time_days (days difference between start_date and withdraw_date)
    try:
        df2['withdraw_date'] = pd.to_datetime(df2['withdraw_date'])
        df2['time_days'] = (df2['withdraw_date'] - pd.to_datetime(df2['start_date'])).dt.days.fillna(0).astype(int)
    except Exception:
        # fallback if parsing fails
        if df2['withdraw_date'].dtype.kind in 'iuf':
            df2['time_days'] = df2['withdraw_date'].astype(int)
        else:
            df2['time_days'] = 365*5
            print("Could not parse withdraw_date; using cap for time_days.")

# ensure numeric types
for col in ['age','salary','tenure','balance_access','balance_retire','time_days']:
    df2[col] = pd.to_numeric(df2[col], errors='coerce').fillna(0)

# event: ensure integer 0/1
df2['event'] = df2['withdraw_event'].astype(int)

# final shape and preview
print("\nPrepared dataset shape:", df2.shape)
print("Preview:")
display(df2.head(5))
```
